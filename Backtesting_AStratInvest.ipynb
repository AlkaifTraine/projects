```python
import pandas as pd
import numpy as np
```

```python
def change_date_format(file_path):
    df = pd.read_csv(file_path)
    df['date'] = pd.to_datetime(df['date'], format='%d-%m-%Y')
    df['date'] = df['date'].dt.strftime('%m-%d-%Y')
    return df
```

```python
def transform_stock_data(df):
    df['date'] = pd.to_datetime(df['date'])
    df['year'] = df['date'].dt.year
    df['month'] = df['date'].dt.month
    grouped = df.groupby(['year', 'month'])
    result = pd.DataFrame()
    result['year'] = grouped['year'].first()
    result['month'] = grouped['month'].first()
    result['open'] = grouped['open'].first()
    result['close'] = grouped['close'].last()
    result['adj close'] = grouped['adj close'].last() 
    result = result.reset_index(drop=True)
    return result
```

```python
for i,stock in enumerate(stocks):
    stocks[i]['prev_month_close']=stocks[i]['adj close'].shift(1)
    stocks[i]=stocks[i].dropna(axis=0,how='any')
    monthly_return=[]
    for j in range(1,25):
        p1=stocks[i].at[j,'adj close']
        p0=stocks[i].at[j,'prev_month_close']
        monthly_return.append((p1/p0)-1)
    stocks[i]['monthly_return']=monthly_return
```

```python
return_stocks=[]
for i,stock in enumerate(stocks):
    df=pd.DataFrame(columns=['year','month','stock','position','buy','sell','profit','monthly_return','cum_return','mean_return','return_volatility','risk_adjusted_return'])
    for j in range(14,25):
        year=stocks[i].at[j,'year']
        month=stocks[i].at[j,'month']
        monthly_return=stocks[i].at[j,'monthly_return']
        p1=stocks[i].at[j-1,'adj close']
        p0=stocks[i].at[j-13,'adj close']
        cum_return=(p1/p0)-1
        sum=0
        sq_diff=0


        
        for k in range(j-12,j):
            sum=sum+stocks[i].at[k,'monthly_return']
        mean_return=sum/12
        for k in range(j-12,j):
            sq_diff=sq_diff+ (stocks[i].at[k,'monthly_return'] - mean_return)**2
        return_volatility=np.sqrt(sq_diff/11)
        risk_adjusted_return=mean_return/return_volatility
        row=pd.DataFrame({'year':[year],'month':[month],'stock':[stock_name[i]],'buy':[stocks[i].at[j,'open']],'sell':[stocks[i].at[j,'adj close']],'monthly_return':[monthly_return],'cum_return':[cum_return],'mean_return':[mean_return],'return_volatility':[return_volatility],'risk_adjusted_return':[risk_adjusted_return]})
        df = pd.concat([df,row], ignore_index=True)
    return_stocks.append(df)
    
```

```python
monthly_sorting=[]
for i in range(len(return_stocks[0])):
    df=pd.DataFrame(columns=['year','month','stock','position','buy','sell','profit','monthly_return','cum_return','mean_return','return_volatility','risk_adjusted_return'])
    for j in range(len(stocks)):
        row = pd.DataFrame({'year': [return_stocks[j].at[i,'year']],'month': [return_stocks[j].at[i,'month']],'stock':[stock_name[j]],'buy':[stocks[j].at[i+14,'open']],'sell':[stocks[j].at[i+14,'adj close']],'monthly_return':[return_stocks[j].at[i,'monthly_return']],'cum_return':[return_stocks[j].at[i,'cum_return']],'mean_return':[return_stocks[j].at[i,'mean_return']],'return_volatility':[return_stocks[j].at[i,'return_volatility']],'risk_adjusted_return':[return_stocks[j].at[i,'risk_adjusted_return']]})
        df = pd.concat([df,row], ignore_index=True)
    df=df.sort_values(by='risk_adjusted_return', ascending=False)
    
    monthly_sorting.append(df)
```

```python
monthly_selected_stocks=[]
extreme=int(len(stocks)*0.2)
for i in range(len(monthly_sorting)):
    buy_index= monthly_sorting[i].head(extreme).index.tolist()
    sell_index=monthly_sorting[i].tail(extreme).index.tolist()
    monthly_selected_stocks.append([buy_index,sell_index])
```

```python
monthly_selected_stocks
```

```python
monthly_weight=[]

for i in range(len(monthly_selected_stocks)):
    
    weight=[]
    for x in range(len(stocks)):
        weight.append(0)
    wb=0.5/len(monthly_selected_stocks[0][0])
    ws=-1*wb
    for j in range(len(monthly_selected_stocks[0])):
        if j==0:
            for k in range(int(len(stocks)*0.2)):
                t=monthly_selected_stocks[i][j][k]
                
                weight[t]=wb
        else:
            for k in range(int(len(stocks)*0.2)):
                t=monthly_selected_stocks[i][j][k]
                
                weight[t]=ws
    monthly_weight.append(weight)           
```

```python
trade_sheet=pd.DataFrame(columns=['year','month','stock','position','buy','sell','profit','monthly_return','cum_return','mean_return','return_volatility','risk_adjusted_return'])

for i in range(len(monthly_sorting)):
    top=monthly_sorting[i].head(int(len(stocks)*0.2))
    bottom=monthly_sorting[i].tail(int(len(stocks)*0.2))
    one_month=pd.concat([top,bottom])
    trade_sheet=pd.concat([trade_sheet,one_month])
```

```python
trade_sheet = trade_sheet.reset_index(drop=True)
trade_sheet.loc[trade_sheet['position'] == 'buy', 'profit'] = (trade_sheet['buy'] - trade_sheet['sell']) / trade_sheet['buy']
trade_sheet.loc[trade_sheet['position'] == 'sell', 'profit'] = (trade_sheet['sell'] - trade_sheet['buy']) / trade_sheet['buy']
```
